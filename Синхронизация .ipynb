<p>$\textbf{Целью}$ данной задачи является изучение синхронизации на основном тоне и обертонах генератора пилообразных колеьаний.</p>

<p>Явление $\textbf{синхронизации}$ заключается в том, что при действии периодической внешней силы на генератор его период изменяется и становится точно кратным периоду внешней силы. </p>

<p>$\textbf{Принцип работы:}$ Наблюдение явления синхронизации осуществляется с помощью осциллографа. На первый и второй входы каналов подаются выходное напряжение релаксатора пилообразной формы и гармонический сигнал, используемый для синхронизации. Т.е. на экране одновременно наблюдается "пила" и синусоида. В общем случае синосоида перемещается относительно неподвижной "пилы" и частота синхронизирующего сигнала некратна частоте "пилы". Если же наступает явление синхронизации, то их частоты становятся кратными. Если наблюдается синхронизация N-го порядка, то на экране 1 период "пилы" соответствует N периодам синусоиды.  </p>

$$\textbf{Упражнение 1}$$
$$\textbf{Исследование зависимости зарядного тока от коллекторного напряжения зарядного транзистора}$$

<p>В данном упражнении был построен график зависимости коллекторного тока зарядного транзистора от коллекторного напряжения. Измерения были проведены при трех произвольных значениях тока базы, величина которого устанавливалась потенциометром R1.
Как видно из графика, существуют участки вольтамперных характеристик, в пределах которых ток коллектора практически не зависит от коллекторного напряжения.  </p>

```python
import matplotlib.pyplot as plt
import numpy as np

u1 = [0, .8, 2.2, 4, 5.2]
i1 = [0, .24, .26, .28, .28]

u2 = [0, .4, 1.6, 2.8, 5.6, 6.2]
i2 = [0, .4, .42, .44, .49, .49]

u3 = [0, .4, 1.2, 2.4, 5.6]
i3 = [0, .67, .7, .73, .81]


plt.ylabel('$Ic, mkA$')  
plt.xlabel('$U, V$')
plt.plot(u1, i1)
plt.scatter(u1, i1)
plt.plot(u2, i2)
plt.scatter(u2, i2)
plt.plot(u3, i3)
plt.scatter(u3, i3)
grid1 = plt.grid(True)
plt.title('Вольтамперная характеристика')
plt.show()
```

<p>В данном упражнении был построен график зависимости частоты релаксационных колебаний от зарядного тока F(I).

$$\textbf{Упражнение 3}$$
$$\textbf{Исследование работы релаксационного генератора в неавтономном режиме}$$

<p>В ходе данного упражнения был определен коэффициент управления K как отношение амплитуды огибающей верхнего уровня пилообразного напряжения к амплитуде звукового генератора.</p>

<p>Далее были исследованы области синхронизации: были сняты зависимости граничных частот областей синхронизации 1, 2 и 3 порядков от амплитуды синхронизирующего напряжения. Результаты представлены на одном графике.</p>
<p>а-амплитуда синхронизирующего сигнала; 
К-коэффициент управления;
V0-амплитула свободных колебаний релаксатора;
fNi-граничные частоты областей синхронизации</p>

```python
K = 2.174
v0 = 5
a = [.7, .889, 1., 1.54, 2., 2.6]
f0 = 8.929
f1_end = [15.51, 16.284, 16.95, 20.092, 23.513, 27.052]
f1_start = [12.444, 11.993, 11.742, 10.826, 10.396, 9.831]
f2_end = [32.46, 33.87, 33.047, 37.584, 44, 52.05]
f2_start = [24.443, 23.256, 23.887, 23.935, 27.683, 31.91]
f3_start = [38.85, 37.515, 37.85, 40.194, 46.932, 59.292]
f3_end = [44.4, 45.8, 47.113, 53.04, 61.254, 74.39]

def norm(lst, f0):
    ans = []
    for elem in lst:
        ans.append(float(f0) / elem)
    return ans

f1 = norm(f1_end, f0)
f1_s = norm(f1_start, f0)
f2 = norm(f2_end, f0)
f2_s = norm(f2_start, f0)
f3 = norm(f3_end, f0)
f3_s = norm(f3_start, f0)

a_new = []
for elem in a:
    a_new.append(K * elem / v0)

plt.ylabel('$Ka/V0$')  
plt.xlabel('$f0/fNi$')
plt.plot(f1, a_new, color='red')
plt.scatter(f1, a_new, color='red')
plt.plot(f1_s, a_new, color='red')
plt.scatter(f1_s, a_new, color='red')
plt.plot(f2, a_new, color='blue')
plt.scatter(f2, a_new, color='blue')
plt.plot(f2_s, a_new, color='blue')
plt.scatter(f2_s, a_new, color='blue')
plt.plot(f3, a_new, color='green')
plt.scatter(f3, a_new, color='green')
plt.plot(f3_s, a_new, color='green')
plt.scatter(f3_s, a_new, color='green')
grid1 = plt.grid(True)
plt.title('')
plt.show()
```

<p>Затем, плавно изменяя амплитуду синхронизирующего сигнала, были найдены пределы изменения амплитуды, при которых наблюдалась синхронизация пятого, четвертого и т.д. порядков. По оси ординат отложены номера порядков синхронизации, по оси абсцисс-амплитуда синхронизирующего сигнала. </p>

```python
y = [5, 5, 4, 4, 4, 3, 3, 3, 2, 2,1, 1, 1]
x = [.2, 1.2, 1.2, 1.32, 2.04, 2.04, 2.44, 2.82, 2.86, 3.48, 3.48, 4.14, 5]
plt.ylabel('$N$')  
plt.xlabel('$A$')
plt.plot(x, y)
plt.scatter(x, y)
grid1 = plt.grid(True)
plt.title('')
plt.show()
```
